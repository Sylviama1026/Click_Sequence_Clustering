```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('user_data.csv',index_col = 'pk_student')

df.head()
```

```python
pd.options.display.float_format = "{:.2f}".format
df.describe()
```

```python
df_active = df.loc[df['Activation'] != 'Extremely active']
df_active.describe()
```

```python
## login_num
plt.figure(dpi=150)
sns.violinplot(x='Activation',y='login_num', data = df_active) 
```

```python
## pv_cnts
plt.figure(dpi=150)
sns.violinplot(x='Activation',y='pv_cnts', data = df_active) 
```

```python
## pv_duration
plt.figure(dpi=150)
sns.violinplot(x='Activation',y='pv_duration', data = df_active) 
```

```python
## click_cnts
plt.figure(dpi=150)
sns.violinplot(x='Activation',y='click_cnts', data = df_active) 
```
